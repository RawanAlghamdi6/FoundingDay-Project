```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
```

```python
file_path = r"C:\\Users\\irawa\\Desktop\\python_ws\\foundingday\\sdaiastudents.csv"
df = pd.read_csv(file_path, encoding='utf-8')
df['sentence'] = df['sentence'].str.replace('\n', ' ')
```

```python
df
```

```python
records = df.to_dict(orient='records')
```

```python
font = ImageFont.truetype("C:\\Users\\irawa\\Desktop\\python_ws\\foundingday\\Harir_complete_OTF_Harir_Bold.otf", size=25)
```

```python
from PIL import Image, ImageDraw, ImageFont
from bidi.algorithm import get_display
import arabic_reshaper

def generate_cards(data):
    template = Image.open("C:\\Users\\irawa\\Desktop\\python_ws\\foundingday\\SaudiFoundingDay.png")
    draw = ImageDraw.Draw(template)

    sentence_width = 0
    sentence_height = 0
    if 'sentence' in data:
        sentence_text = data['sentence']
        sentence_font_size = 190
        sentence_font = ImageFont.truetype("C:\\Users\\irawa\\Desktop\\python_ws\\foundingday\\Harir_complete_OTF_Harir_Bold.otf", sentence_font_size)
        reshaped_sentence = arabic_reshaper.reshape(sentence_text)
        bidi_sentence = get_display(reshaped_sentence)
        sentence_position = ((template.width - sentence_width) // 2 - 250, (template.height - sentence_height) // 2 - 150)
        draw.text(sentence_position, bidi_sentence, font=sentence_font, fill='black')
        
    fullname_width = 0
    fullname_height = 0
    if 'fullname' in data:
        fullname_text = data['fullname']
        fullname_font_size = 170
        fullname_font = ImageFont.truetype("C:\\Users\\irawa\\Desktop\\python_ws\\foundingday\\Harir_complete_OTF_Harir_Bold.otf", fullname_font_size)
        reshaped_fullname = arabic_reshaper.reshape(fullname_text)
        bidi_fullname = get_display(reshaped_fullname)
        fullname_position = ((template.width - fullname_width) // 2, (template.height - fullname_height) // 2 + 150)
        draw.text(fullname_position, bidi_fullname, font=fullname_font, fill='black')
        
    return template
```

```python
for record in records:
    card = generate_cards(record)
    if card is not None:
        # Save the image to a file
        card.save(f"cards/{record['id']}.png")
```
